## The attention distribution

Let's get the attention distribution for the first block. We'll use the manhattan metric to compute the match-scores.

```python
embeddings = torch.nn.Embedding.from_pretrained(model.state_dict().get("embeddings.weight"))
```

```python
x1_0 = embeddings(batch[:, 0, :]).unsqueeze(1)
x2_0 = embeddings(batch[:, 1, :]).unsqueeze(1)
```

```python
A_0 = compute_attention_matrix(x1_0, x2_0, manhattan)
A_0 = A_0.squeeze().numpy()
```

```python
def generate_tick_labels(q1, q2, max_length=20):
    """ Creates labels for each tick on the heatmap. 
    
        Args:
            q1, q2: strings
                The text in the question pair.
            max_length: int
                The maximum length for any sequence.
                
        Returns:
            xticks, yticks: lists of strings
                The tokenized sequences. Padding tokens are appended
                if the sequences are contain fewer than max_length tokens.
    """
    # Sanity checks
    assert(isinstance(q1, str) and isinstance(q2, str) and isinstance(max_length, int))
    
    # Convert text to list of words, just as is done by the model
    words1 = text_to_word_list(q1)
    words2 = text_to_word_list(q2)   

    # Remove stop words that are not in the word2vec vocabulary
    words1 = list(filter(lambda s: s not in stops or s in word2vec.vocab, words1))
    words2 = list(filter(lambda s: s not in stops or s in word2vec.vocab, words2))
    
    # Make the labels
    yticks = list(words1)
    yticks = yticks + ["<unk>" for _ in range(max_length - len(yticks))]
    xticks = list(words2)
    xticks = xticks + ["<unk>" for _ in range(max_length - len(xticks))]
    
    return xticks, yticks
```

```python
def generate_heat_map(A, q1, q2):
    """ Generates a heat map for the attention matrix. 
    
        Args:
            A: np.ndarray of shape (max_length, max_length)
                The attention matrix
            q1, q2: strings
                The text in the question pair
        
        Returns:
            None
    """
    # Sanity checks
    assert(len(A.shape) == 2 and A.shape[0] == A.shape[1])
    assert(isinstance(q1, str) and isinstance(q2, str))
    
    # Make plot bigger so its easier to read
    plt.rcParams['figure.figsize'] = 10, 10

    # Make custom x and y ticks with labels for this example
    max_length = A.shape[0]
    xticks, yticks = generate_tick_labels(q1, q2, max_length=max_length)

    # Make x ticks and labels appear on top
    plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
    plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True

    # Plot the attention distribution
    plt.imshow(A, cmap='hot', interpolation='nearest')
    plt.xticks(range(20), xticks, rotation=90)
    plt.yticks(range(20), yticks)
    plt.colorbar()
    plt.show()
```

```python
generate_heat_map(A_0[0], examples[0][0], examples[0][1])
```

```python
generate_heat_map(A_0[1], examples[1][0], examples[1][1])
```

```python
generate_heat_map(A_0[2], examples[2][0], examples[2][1])
```

Now let's try seeing what the attention distributions look like after the input goes through the first ABCNN-3 Block. First we'll need to replicate the architecture of the first Block.

```python
# Initialize the sub-modules for the ABCNN-3 Block
attn1_1 = ABCNN1Attention(300, 20, True, "manhattan")
conv_1 = Convolution(300, 50, 3, 2)
attn2_1 = ABCNN2Attention(20, 3, "manhattan")

# Load the pre-trained weights
attn1_1.W1.data = model.state_dict().get("blocks.0.attn1.W1")
conv_1.conv.weight.data = model.state_dict().get("blocks.0.conv.conv.weight")
conv_1.conv.bias.data = model.state_dict().get("blocks.0.conv.conv.bias")

# Create the first ABCNN-3 Block
block_1 = ABCNN3Block(attn1_1, conv_1, attn2_1)

# Process the inputs
x1_1, x2_1, a1_1, a2_1 = block_1(x1_0, x1_0)

# Create the attention matrix
A_1 = compute_attention_matrix(x1_1, x2_1, manhattan)
A_1 = A_1.squeeze().detach().numpy()
```

```python
generate_heat_map(A_1[0], examples[0][0], examples[0][1])
```

```python
generate_heat_map(A_1[1], examples[1][0], examples[1][1])
```

```python
generate_heat_map(A_1[2], examples[2][0], examples[2][1])
```

And now let's see what the attention distribution looks like after the input passes through the second Block.

```python
# Initialize the sub-modules for the ABCNN-3 Block
attn1_2 = ABCNN1Attention(300, 20, True, "manhattan")
conv_2 = Convolution(300, 50, 3, 2)
attn2_2 = ABCNN2Attention(20, 3, "manhattan")

# Load the pre-trained weights
attn1_2.W1.data = model.state_dict().get("blocks.1.attn1.W1")
conv_2.conv.weight.data = model.state_dict().get("blocks.1.conv.conv.weight")
conv_2.conv.bias.data = model.state_dict().get("blocks.1.conv.conv.bias")

# Create the first ABCNN-3 Block
block_2 = ABCNN3Block(attn1_2, conv_2, attn2_2)

# Process the inputs
x1_2, x2_2, a1_2, a2_2 = block_2(x1_1, x1_1)

# Create the attention matrix
A_2 = compute_attention_matrix(x1_2, x2_2, manhattan)
A_2 = A_2.squeeze().detach().numpy()
```

```python
generate_heat_map(A_2[0], examples[0][0], examples[0][1])
```

```python
generate_heat_map(A_2[1], examples[1][0], examples[1][1])
```

```python
generate_heat_map(A_2[2], examples[2][0], examples[2][1])
```
